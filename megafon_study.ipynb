```python
metrics = groups.columns[2:10]
start_metric = metrics[0]
print(start_metric)
metrics
```

```python
metrics_names = ['объем трафика передачи данных', 'средняя скорость «к абоненту»', 'средняя скорость «от абонента»', 
                 'частота переотправок пакетов «к абоненту»', 'скорость загрузки потокового видео', 
                 'задержка старта воспроизведения видео', 'скорость загрузки web-страниц через браузер',
                 'пинг при просмотре web-страниц'] 
```

```python
import plotly.graph_objects as go
fig = go.Figure()
for metric in metrics:
    visible = False
    if metric == start_metric: 
        visible = True 
    for group_num, color in list(zip([1, 2, 3], ['lightblue', 'lightgreen', 'pink'])):
        quant = groups[metric].quantile(q = 0.95)
        datas = groups.loc[(groups['group'] == group_num) & (groups[metric] < quant)][metric]
        fig.add_trace(
            go.Histogram(
                x=datas,
                marker_color=color,
                visible=visible,
                name='группа ' + str(group_num))
        )
        

mysteps = []
for i in range(len(metrics)):
    tmp = []
    for j in range(len(metrics)):
        if (i == j):
            tmp.append(True)
        else:
            tmp.append(False)
    step = dict(
        method="update",
        args = [
            {'visible': tmp},
            {'title.text' : 'Распределение показателей по метрике {} у групп пользователей'.format(metrics_names[i])}
        ],
        label = metrics[i][:10]
    )
    mysteps.append(step)
    

fig.update_layout(
    sliders=[dict(
        steps=mysteps, 
    )]
)

fig.update_layout(
    title={'text': 'Распределение показателей по метрике {}'.format(metrics_names[0])}
)       

fig.write_html("compare_metrics.html")
fig.show()    
```

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
groups = pd.read_csv('megafonGroups.csv')
groups.head()
```

```python
from scipy.stats import skew
print('Коэффициенты ассиметрии для распределений метрик связи')
print('')
for col in groups.columns[2:-1]: 
    print(str(col) + '= ' + str(skew(groups[col])))
```

Проверяю распределения для каждой метрики. Они все ненормальные, скошенные влево. <br> Поэтому я не могу сравнивать доверительных интервалы этих групп, чтобы понять, пересекаются они или нет? <br> Поэтому просто взять непараметрический тест? 
<br> Возьму тест Краскела-Уоллиса, для более, чем двух выборок и при ненормально распределении. 

```python
for metric in list(groups.columns[2:-1]): 
    quant = groups[metric].quantile(q=0.95)
    fig, ax = plt.subplots(figsize=(6, 4))
    datas = groups.loc[groups[metric] < quant, metric]
    sns.histplot(datas, ax = ax)
```

Использую новую библиотеку для стат.анализа, чтобы посмотреть результаты теста Краскела-Уоллиса для каждой метрики связи - отличаются ли их показатели у трех групп? 

```python
import pingouin as pg
kruskal_stats = pd.DataFrame()
for metric in list(groups.columns)[:-1]:
    kruskal_tests = pg.kruskal(data=groups, dv='group', between=metric)
    kruskal_stats = kruskal_stats.append(kruskal_tests)
kruskal_stats
```

Результаты будто говорят, что только у двух метрик есть стат.значимые различия: Web Average TCP RTT(ms) (р<0.01), 
<br>Downlink TCP Retransmission Rate(%)(p<0.002)
* т.е. у третьей группы (кто поставил ниже 4 баллов за качество связи), более высокие показатели по Downlink TCP Retransmission Rate(%)) и Web Average TCP RTT(ms). 
#### Стоит обратить на это внимание, поэтому что, видимо, у этих пользователей действительно другие показатели по этим метрикам (стат.значимы)

##### Почему бутстреп? 
Потому что распределения по каждой метрике - ненормальные. Бутстреп поможет симулировать нормальное распределение средних по множествам симулированных выборок.
